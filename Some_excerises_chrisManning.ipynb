```python
# import modules
import nltk
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Downloading files from Gutenberg
from urllib.request import urlopen 
url ="http://www.gutenberg.org/files/2554/2554.txt"
raw =urlopen(url).read()
type(raw)
```

```python
raw
```

```python
import string, math 
alpha_list =list(string.ascii_lowercase)
list_alprobs=[]
sum1=0
for alphabet in alpha_list:
        sum1+= str(raw).count(alphabet)/len(raw) * math.log(str(raw).count(alphabet)/len(raw), 2) 
        list_alprobs.append(str(raw).count(alphabet)/len(str(raw)))
print(sum1)        
        
```

```python
plt.plot(list_alprobs)
plt.grid(True)
plt.xlabel('Alphabet')
plt.ylabel('Probability of Alphabet')
plt.title('Probability of Alphabet in a specified text')
plt.annotate('Entropy:-3.209588 bits', xy= (10, 0.06)) # no nneed to specify the xytext parameter.
plt.show()
```

```python
char_list = set(str(raw))
sum2=0
list_probs= []
for char in char_list:
        sum2 += str(raw).count(char)/len(str(raw)) * math.log(str(raw).count(char)/len(str(raw)), 2)
        list_probs.append(str(raw).count(char)/len(str(raw)))
print(sum2)
        
```

```python
plt.plot(list_probs, 'b')
plt.grid(True)
plt.xlabel("characters")
plt.ylabel('Probability of Character')
plt.title('Probability of Occurance of character in specified text')
plt.annotate('Entropy: -5.0022 bits',xy=(35, 0.12) , xytext=(35, 0.12))
plt.show()
```
